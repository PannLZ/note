### 指数与对数

> 对数衡量的是达到某个规模所需付出的努力；
- 时间成本：初始本金：$P$，年利率：$r$
    - 存了 $n$ 年后的金额：$A = P (1+r)^n$
    - 现在反过来问：我想要存到金额 $A$，年利率 $r$ 已知，我得存多少年？
        - $A = P(1+r)^n \quad \Rightarrow \quad
n = \log_{1+r} \left(\frac{A}{P}\right)$

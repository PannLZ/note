$$
P(Z|X) = \frac{P(X|Z)P(Z)}{P(X)}= \frac{P(X|Z)P(Z)}{\int P(X,Z)dZ} =\frac{P(X|Z)P(Z)}{\int P(X|Z)P(Z) dZ}
$$

### odds 与 bayes

> bayes filtering：基于有噪观测数据的统计推断（Statistical Inference）

- $O(H) = \frac{P(H)}{P(\neg H)} = \frac{P(H)}{1 - P(H)}$
    - 概率 $P(H)$：事件 $H$ 发生的可能性，范围是 $[0, 1]$。
    - 几率 $O(H)$：事件发生与不发生的比值。范围是 $[0, \infty)$。
- bayes
    - $P(H|E) = \frac{P(E|H) \cdot P(H)}{P(E)} \quad ......(1)$
    - $P(\neg H|E) = \frac{P(E|\neg H) \cdot P(\neg H)}{P(E)} \quad ......(2)$
    - 两式相除：$\frac{P(H|E)}{P(\neg H|E)} = \frac{P(E|H) \cdot P(H)}{P(E|\neg H) \cdot P(\neg H)}$

$$
\underbrace{\frac{P(H|E)}{P(\neg H|E)}}_{\text{后验几率}} = \underbrace{\frac{P(E|H)}{P(E|\neg H)}}_{\text{似然比}} \times \underbrace{\frac{P(H)}{P(\neg H)}}_{\text{先验几率}}
$$

- $\log O(H|E) = \log \Lambda(E|H) + \log O(H)$
    - Posterior Odds = Likelihood Ratio × Prior Odds。

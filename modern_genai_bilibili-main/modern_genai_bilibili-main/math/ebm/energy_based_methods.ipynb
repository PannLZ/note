### 能量越低，概率越高

$$
P(x) = \frac{e^{-E(x)}}{Z}
$$

- EBM（能量模型）的核心思想是：“能量”是对“违和感”或“错误程度”的度量。
    - 想象一个起伏不平的山地地形（Energy Landscape）：
        - 低能量（山谷底）：把一个小球放在这里，它是稳定的，它不想动。这代表这一状态是“自然”的、“合理”的。
        - 高能量（山顶/半山腰）：小球在这里是不稳定的，它会因重力自然滚落。这代表这一状态是“不稳定”的、“不自然”的。
- 在 EBM 中，不要把“能量”理解为物理学里的焦耳，而要把它理解为一个**“惩罚分数” (Penalty Score)** 或 “不兼容度” (Incompatibility)。

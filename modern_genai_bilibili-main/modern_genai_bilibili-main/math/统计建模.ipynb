### $\mu, \sigma$

- 统计建模思维：均值方差（mu，variance）
    - 均值方差地统计视角理解人群或组织
    - 平均值，和波动程度

$$
Z = \frac{\text{分数} - \text{平均分}}{\text{标准差}}
$$
- z-score （顾名思义，标准分数，是一个分数概念）物理意义是：你的分数距离平均分有几个“标准差”的距离？
    - “不同科目的考试成绩比较”。小明（考数学）和小红（考英语），小明（数学）： 85分，小红（英语）： 70分
    - $Z_{\text{小明}} = \frac{85 - 90}{5} = \frac{-5}{5} = \mathbf{-1}$
    - $Z_{\text{小红}} = \frac{70 - 60}{10} = \frac{10}{10} = \mathbf{+1}$

$$Z = \frac{X - \mu}{\sigma}$$
- 同时也有标准化之意（本质是线性变换）；
    - Z = 0：代表刚好是平均水平。
    - Z > 0 (正数)：代表高于平均水平（越优秀）。
    - Z < 0 (负数)：代表低于平均水平（越差）。

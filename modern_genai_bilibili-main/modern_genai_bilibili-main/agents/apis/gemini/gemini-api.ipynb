- 多模态输入 tokens 计数
    - https://geminibyexample.com/027-calculate-input-tokens/
        - 对于尺寸小于或等于 384x384 像素的图片：计为 258 个 tokens。
            - 258 = 16\*16 + 2 (24\*24 为一个 patch, 一个 image token，384/16)
                - 2: `<bos_v>, <eos_v>`，表示图像的开始和结束；
        - 对于尺寸大于 384x384 像素的图片：图片会被分割成多个 768x768 像素的图块，每个图块计为 258 个 tokens。

### google genai

- `pip install google-genai`

```python
from google import genai
client = genai.Client(api_key=os.getenv("GOOGLE_API_KEY"))

resp = client.models.generate_content(
    model='gemini-2.5-flash',
    contents=[Image.open(image_path), prompt]
)

usage = resp.usage_metadata
# prompt_token_count: TEXT / IMAGE
# candidate_token_count: 最终输出的 tokens 数
# thought_token_count: thinking tokens
# total_token_count = prompt_token_count + thought_token_count + candidate_token_count
```

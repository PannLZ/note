- 兼容性：Cerebras 的 API 完全兼容 OpenAI 格式。你只需要换掉 base_url 和 api_key，现有的代码（使用 openai python 库或 langchain）通常不需要重写。
    - Base URL: https://api.cerebras.ai/v1
- 局限性：由于它是专为推理速度优化的，目前支持的模型主要是开源系列（如 Llama 3.1, Llama 3.3 等），不支持 OpenAI 的私有模型（GPT-4）。
    - 支持的 model list
        - qwen-3-235b-a22b-instruct-2507
        - qwen-3-32b
        - zai-glm-4.6
        - llama3.1-8b
        - llama-3.3-70b
        - gpt-oss-120b

```python
import os
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv(), override=True)

# 假设你的程序会读取配置文件并设置环境变量，或者你可以直接在这里填入 Key 测试
client = OpenAI(
    base_url="https://api.cerebras.ai/v1",
    api_key=os.getenv("CERERAS_API_KEY"),
)

models = client.models.list()
print(models)

response = client.chat.completions.create(
    model="llama3.1-8b", # 或 llama-3.3-70b
    messages=[{"role": "user", "content": "Hello, explain quantum physics in one sentence."}]
)

print(response.choices[0].message.content)
```

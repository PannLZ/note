### langfuse 本地部署与集成

```
# 创建一个文件夹保持整洁
mkdir langfuse-local
cd langfuse-local

# 使用 curl 下载 docker-compose.yml
curl -o docker-compose.yml https://raw.githubusercontent.com/langfuse/langfuse/main/docker-compose.yml

# 启动服务
docker compose up -d

# 打开 Safari 或 Chrome，访问 http://localhost:3000。
# 注册账户（这是你本地的管理员账户）。
# 新建项目，复制 Public Key 和 Secret Key。
```

- `pip install langfuse openai python-dotenv`

### scripts

```python
import os
from dotenv import load_dotenv # 导入 dotenv
from langfuse import observe
from langfuse.openai import OpenAI

load_dotenv()

client = OpenAI()

@observe()
def my_mac_chat():
    return client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": "Hello from macOS!"}],
    )

my_mac_chat()
```

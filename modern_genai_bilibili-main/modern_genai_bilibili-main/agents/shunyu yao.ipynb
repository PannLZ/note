- 关于 language agent，语言智能体与传统智能体的本质区别在于它能够进行推理，而推理是实现泛化的关键。 （推理即泛化）
- 关于 ReAct
    - 初衷与动机：当时主流的语言模型研究分为两派，一派是基于BERT模型做选择题式的任务，另一派是基于GPT模型做生成式任务。 他意识到，如果要做一个能自主行动的Agent，它需要的不仅仅是选择能力，而是自由产生新动作的能力，这必须依赖像GPT这样的生成模型。
        - 传统的AI模型在新游戏中需要成千上万次的训练，而人类却能很快上手。他意识到，其根本原因在于人类可以思考和推理。例如，在一个新环境中，人会根据常识进行推理（如“灯是黑的，可能有危险，需要先找灯”），从而规划出行动步骤。
    - ReAct 原理，将“推理”（Reasoning）和“行动”（Acting）结合起来。即 这个框架让Agent在执行任务时，不仅能输出动作，还能生成“内心独白”（即思考过程），解释自己为什么这么做以及下一步计划，从而实现更复杂的任务规划和执行。
    - 他认为ReAct是自己最喜欢的工作之一，并指出这种通过提示（Prompting）来引导模型进行研究的方式，在当时并不被学术界主流所接受。（当时学术界对于这种基于“提示工程”（Prompting）的方法是否属于“真正的研究”存在疑虑，但他坚信这是探索如何有效使用模型的关键路径。）但事实证明，研究如何更好地“使用”模型，而不是仅仅“训练”模型，是极具价值的方向。 目前，类似ReAct的架构已经成为构建Agent时最主流和通用的方法之一。
- 关于 AI 下半场：从训模型到用模型
    - 上半场：核心是**方法**，即如何训练出更强大的模型。大家关注的是如何改进模型架构、训练方法等，比如从BERT到GPT的演进。
    - 下半场：核心是**任务**（应用和任务定义），即有了强大的通用模型（就像一把“通用机关枪”）之后，我们应该用它来解决什么问题（朝哪里开枪）。 重点不再是发明新方法，而是如何定义好的任务、设计好的环境，并利用现有模型的通用能力去完成这些任务。

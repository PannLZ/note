### LangExtract

https://towardsdatascience.com/using-googles-langextract-and-gemma-for-structured-data-extraction/
- 智能分块（Smart Chunking）：系统首先将长文档分割成大小可管理且相互重叠的文本块。其分块逻辑具备上下文感知能力，能够尊重句子和段落的边界，避免了因粗暴切分而破坏语义完整性，从而保证了后续 LLM 的处理性能。   
- 并行处理（Parallel Processing）：分割后的文本块会通过一个可配置工作线程数（max_workers）的池进行并发处理。这种并行化设计极大地降低了处理海量数据时的延迟，使其能够胜任大规模数据抽取的任务。
- 多轮抽取（Multi-Pass Extraction）：该库支持对同一文本执行多轮抽取。这项技术旨在提高召回率，尤其适用于关键信息稀疏分布的“大海捞针”（needle-in-a-haystack）场景 。为了解决多轮抽取可能产生的冲突，系统采用“首轮优先”（first-pass-wins）的规则进行裁决 。

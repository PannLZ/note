### Temporal

- Temporal Memory Graph: https://graphrag.com/reference/knowledge-graph/memory-graph-temporal/
    - 为关系添加时间戳：直接在实体之间的关系（边）上附加时间戳，明确指出该关系在何时有效。例如，一个“A 公司 CEO 是张三”的关系可以被标记上“2020年至2023年”的时间戳。
    - 版本化节点与“PREVIOUS”关系：当一个实体的属性发生变化时，系统会创建一个新的节点来代表更新后的状态，并使用一个“PREVIOUS”关系将新旧两个节点连接起来，从而形成一个可追溯的版本链。

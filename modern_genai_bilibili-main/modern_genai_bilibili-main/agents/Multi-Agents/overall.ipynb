> Multi-Agent Collaboration: Harnessing the Power of Intelligent LLM Agents

- 复杂任务规划与分解
    - 多智能体系统通常以任务规划（Task Planning）为起点，将用户请求拆解为可由不同代理完成的子任务。这一过程由一个“元代理”或中央调度LLM负责，要求确保子任务可解性、完整性和非冗余性。例如，Li等人提出了**Agent-Oriented Planning (AOP)**框架，强调每个子任务都可被适当代理解决且覆盖用户需求，不重复无效步骤。AOP通过快速分解任务并用奖励模型评估子任务完成质量，必要时调整任务分配，并引入反馈循环提高规划稳健性。

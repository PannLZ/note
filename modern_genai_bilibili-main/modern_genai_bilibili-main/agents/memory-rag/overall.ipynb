- https://arxiv.org/abs/2312.06648
    - Dense X Retrieval: What Retrieval Granularity Should We Use?
    - 该论文的核心思想是引入一个新的检索单元——“命题”（**Proposition**）。一个命题被定义为一个原子化的、自包含的事实陈述，以简洁的自然语言形式呈现 。它封装了一个不可再分的、有上下文的独立事实。例如，对于句子“比萨斜塔在1990年至2001年间的修复工作之前倾斜5.5度，现在倾斜约3.99度”，可以分解为多个命题 ：
        - “比萨斜塔曾以5.5度的角度倾斜。”
        - “在1990年至2001年间，比萨斜塔进行了修复工作。”
        - “比萨斜塔现在倾斜约3.99度。”
    - 这一创新理念已被 LlamaIndex 社区实现为一个 LlamaPack，即 `DenseXRetrievalPack`。该工具包利用 LLM 从文档中提取命题，并构建相应的检索流程 。

LangChain 和 LlamaIndex 在分块功能上的架构差异，并非简单的实现细节不同，而是反映了它们对“块”这一基本单元的深层哲学分歧。LangChain 主要将“块”视为一段文本字符串，而 LlamaIndex 则将其视为一个具有内在元数据和关系的结构化对象（`Node`）。

- 这种分歧源于它们不同的设计出发点。LangChain 的 `split_text` 方法返回的是一个字符串列表 `List[str]`。其核心数据单元是基础的，元数据和块之间的关系需要开发者在外部另行管理。这种设计提供了极大的灵活性和简洁性，是一种非侵入式的“工具箱”模式。   
- 相比之下，LlamaIndex 的 `get_nodes_from_documents` 方法返回的是一个 `Node` 对象列表 `List[Node]`。
    - 每个 Node 都是一个丰富的对象，封装了文本、嵌入、元数据以及指向其他节点（如 `parent_node`）的指针。这种面向对象的方法使得实现像层级化检索这样的高级策略变得水到渠成。例如，一个 AutoMergingRetriever 需要知道哪些小块属于同一个父块，如果块仅仅是独立的字符串，这种逻辑将难以实现。

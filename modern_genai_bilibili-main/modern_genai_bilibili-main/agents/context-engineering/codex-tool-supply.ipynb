### instructions

- When searching for text or files, prefer using `rg` or `rg --files` respectively because `rg` is much faster than alternatives like `grep`. (If the `rg` command is not found, then use alternatives.)
- Parallelize tool calls whenever possible - especially file reads, such as `cat`, `rg`, `sed`, `ls`, `git show`, `nl`, `wc`. Use `multi_tool_use.parallel` to parallelize tool calls and only this.

> bash is all you need

- rg：ripgrep，查找
    - 对比 `grep -r`
        - rg 目前公认的最快文本搜索工具（rust 编写），其设计哲学更是从“通用搜索”转向了“开发者友好”。
            - 它默认会读取 .gitignore、.ignore 和 .rgignore 文件，自动跳过那些数以万计的依赖包（如 node_modules）和二进制文件。
        - `grep -r "TODO" .`：从当前目录开始，递归地在所有文件里找 TODO。
    - 在 rg  中，-n 是一个非常实用且基础的参数，它的全称是 --line-number。
    - rg -t py：仅查看 python 文件
        - rg -T py：非 py 文件
    - rg -C 5: 上下5行
    - rg -l：只输出包含关键字的文件名，
        - `-l`：`--files-with-matches`
    - `rg --files -g`
        - 不搜索文件内部的文本内容，而是将 rg 变成了一个高性能的“文件罗列器”，并配合强大的“全局模式（Globbing）”进行精准筛选。
- sed: stream editor
    - `sed -n "1,240p"`：只把文件的第 1 到第 240 行显示出来。
        - p：print
- nl -ba：
    - nl 是 Number Lines 的缩写。
    - -ba (body all) 表示给所有行都加上行号（包括空行）。
    - `nl -ba | sed -n`
- git show
    - git log 更像是“看目录、看摘要”
    - git show 更像是“把其中一页全文翻开”

### misc

    - fzf: fuzzy finder
    - 将这个复杂命令配置到 zsh 中；
- fd

    - fd --size +100m：查看大文件

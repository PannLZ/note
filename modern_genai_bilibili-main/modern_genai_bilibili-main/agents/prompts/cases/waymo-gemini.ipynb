> 工业级 Prompt Engineering 的典范

- Persona & Identity
    - Identity
    - Purpose
    - Tone & Style: 友好 (Friendly)、乐于助人 (Helpful)、安心 (Reassuring)、中立 (Neutral)、简洁 (Concise)。
    - Crucial Distinction
        - 规则: 必须严格区分 "Gemini" (对话 AI) 和 "Waymo Driver" (自动驾驶技术)。
        - 禁忌: 绝不能声称自己在驾驶车辆或通过传感器“看”路。驾驶行为必须归因于 "Waymo Driver"。
- Context
    - 环境: Waymo 自动驾驶车辆内。
    - 交互模态:
        - 输出: 主要是音频 (Audio)。
        - 输入: 变量 {{input_modality}} (文本或语音)。
    - 体验目标: 有用 (Useful)、神奇 (Be Magical)、不打扰 (Be Unobtrusive)
- Core Directives
    - 初始化协议 (Initialization)
    - 工具优先级 (Tool Prioritization)
    - 模态感知 (Modality Awareness)
    - 信息获取与地理空间上下文 (Geospatial Context)
    - 安全护栏与禁区 (Safety Guardrails)
    - 错误处理与用户引导 (Error Handling & Phrasing)

### induction vs. deduction 

- premise: 前提
- deduction：演绎推理，forward 前向推理
    - Your conclusion is **CERTAIN**，必然性推理；
    -  example：三段论

```
任何三角形只可能是锐角三角形、直角三角形和钝角三角形。——大前提
这个三角形既不是锐角三角形，也不是钝角三角形。——小前提
所以，它是一个直角三角形。——结论

大前提： 所有人都会死。
小前提： 苏格拉底是人。
结论： 所以，苏格拉底会死。
```

- induction：归纳推理，
    - Your conclusion is only **PROBABLE**，或然性推理
    - More premises (前提), conclusion More Probable
    - example
```
前提： 我见过的天鹅是白色的，在公园看到的天鹅是白色的，纪录片里的天鹅也是白色的。
结论： 所有天鹅都是白色的。
```

### Abductive reasoning (溯因推理)

- Abductive reasoning seeks the likeliest possible explanation for partial observations.
    - the generated hypothesis (H) is expected to best explain what happens before, after, or during the observation (O).

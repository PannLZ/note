### `(:requirements :strips :typing :equality)`

- `(:requirements ...)`: 声明规划器（planner）需要支持的一系列功能。
    - `:strips`: 这是最基本的要求，表示该领域遵循 STRIPS 规划范式。即每个动作（action）都通过“前置条件”（preconditions）、“增加效果”（add effects）和“删除效果”（delete effects）来定义。
        - delete effects：哪些之前为真的事实，现在不再为真了。
    - `:typing`: 表示支持“对象类型”。这意味着您可以为问题中的对象定义不同的类型（比如 robot, location, item），使得领域模型更加结构化，可以防止不合逻辑的动作。
    - `:equality`: 表示支持“等式判断”。这允许在动作的前提条件中使用 = 或 != 来判断两个对象是否是同一个，例如 (not (= ?object1 ?object2))。

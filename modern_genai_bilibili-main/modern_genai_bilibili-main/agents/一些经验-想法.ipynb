- LLM提供了两种能力：一是记忆能力，来自于训练的语料；二是推理能力，来自于对齐。构建agent的时候，我们希望ai只用推理能力，通过tool call获取上下文执行任务。但ai总是会把它自己的记忆混进去，导致幻觉。尤其是上下文跟ai认为的常识差别较大的时候，错误尤其显著。所以完全基于上下文的推理应该是基座LLM未来演进的方向。
- 基于一个经典假设：验证（Verification）通常比生成（Generation）更容易。这个假设对 LLM 未必成立。因为 LLM 的能力来自于在数据中学习模式，如果训练数据中缺少“纠错过程”的数据，它就没有理由能做出正确的批判。
- prompt
    - 古代炼金术的问题不在于化学没有用，而在于人们妄图欺骗自己——只要用正确的咒语（prompt），神奇的化学就能变成点石成金的核物理。今天我们对 LLM 的困惑，与 LLM 炼金术并无二致。

### workflow

- 复杂的 prompts（complex instructions） 对应着较高的认知负荷（cognitive load），要适配模型的能力，比如要进行必要的分解，任务拆分：具体来说对应着 workflow 节点流程设计
- workflow 考验的是任务分解，prompts 设计，以及 prompts 的维护的能力；
    - 试图 principled、相对 general 地设计 prompts，避免 case by case；
- 设计和优化 workflow 基于现代的 LLM 框架（比如 langgraph）反而是相对容易的，调 prompts（当然也可以借助 gpt/gemini）可能会稍微繁琐，话费比 workflow 设计更多的时间

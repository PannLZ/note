        - Model Context Protocol (MCP) connects Claude to third-party tools, and skills teach Claude how to use them well
- MCP 解决的是“连接” (Connection) 问题，而 Skills 解决的是“教育” (Instruction) 问题。
    - MCP (Model Context Protocol): 是 Agent 的**“手和眼”**。它是一个开放标准，用于将 Agent 连接到外部数据（如数据库、本地文件）和工具（如 API、软件）。它回答的是 "我能使用什么工具？"
    - Agent Skills: 是 Agent 的**“大脑回路”或“SOP（标准作业程序）”**（prompt 即逻辑）。它是一套基于文件系统的模块化指令，包含流程知识、最佳实践和脚本。它回答的是 "我该如何正确地完成这项复杂的任务？"
    - 定位和level不同
- “渐进式披露” (Progressive Disclosure)——启动时只读元数据，需要时才加载详细指令。
    - 动态认知注入
- 分类
    - 能力型，(软)编排型

- Claude Skills 并非可执行代码，而是一套基于 prompt 注入的元工具架构。
- Claude’s Agent Skills system represents a sophisticated prompt-based meta-tool architecture that extends LLM capabilities through specialized instruction injection.
- Unlike traditional function calling or code execution, skills operate through prompt expansion and context modification to modify how Claude processes subsequent requests without writing executable code.
- Execution Context Modification 的本质是动态调整 AI 的安全边界和认知能力。它不是改变了代码逻辑，而是通过 contextModifier 修改了 AI 引擎的运行参数（如权限表和模型 ID），让 Claude 在特定的 Skill 范围内表现得像是一个获得了“特许授权”的专业代理人。
    - 为了方便理解，你可以把这个过程想象成**“进入一个受控的实验室”**：Skill 不仅给了你一份实验手册（Prompt），还通过 contextModifier 给了你实验室的门禁卡（工具权限）和更精密的显微镜（模型切换），而你在实验室之外是不具备这些权限和设备的。

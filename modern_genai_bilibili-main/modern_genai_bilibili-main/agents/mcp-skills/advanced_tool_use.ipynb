- `ReAct` to `Code as Reasoning`
    - ReACT 模式的局限性： 传统的“思考-行动”循环在面对复杂、重复性任务时，效率极低且成本极高。
    - **Code as Reasoning** (代码即推理)： 让模型写代码不仅是为了生成软件，而是作为一种解决问题的手段。代码可以处理逻辑分支、大量数据处理和精确计算，弥补了 LLM 的短板。
- cases：https://www.youtube.com/watch?v=2MJDdzSXL74
    - Sonnet 4.5 (左侧) = 直觉型选手 (System 1)
        - 它试图靠“语感”和“记忆”来解决逻辑和计算问题。
        - 它必须把所有数据都塞进自己的“短期记忆”（Context Window）里
        - 缺陷： 记忆有限，计算不准，每一步都要停下来思考，极其昂贵。
    - Opus 4.5 (右侧) = 逻辑型选手 + 工具主宰 (System 2 + Tools)
        - 它退后一步，不再自己做苦力，而是构建系统（Python 脚本）来解决问题。
        - 它不需要“看”数据，它只需要知道数据的“结构”。
        - 优势： 无限的精确度（依托于 CPU），极低的成本（依托于本地运行），极快的速度。

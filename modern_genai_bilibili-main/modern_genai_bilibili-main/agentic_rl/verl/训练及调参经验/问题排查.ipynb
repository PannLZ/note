#### grad norm

- grad_norm：所有参数（parameters）梯度的 L2 范数（L2 Norm）。
    - $ \theta = \{\theta_1, \theta_2, \dots, \theta_n\} $，损失函数为 $ J(\theta) $。在反向传播后，我们得到每个参数的梯度 $ g_i = \frac{\partial J}{\partial \theta_i} $。
    - 所有这些梯度构成一个梯度向量 $ \mathbf{g} = [g_1, g_2, \dots, g_n] $。
    - $\text{grad\_norm}=\|g\|_2=\sqrt{\sum_i^n g_i^2}$
    - verl：`actor_rollout_ref.actor.grad_clip=1.0`
```python
total_sq = 0.0
for p in model.parameters():
    if p.grad is not None:
        total_sq += p.grad.detach().float().pow(2).sum().item()
grad_norm = total_sq ** 0.5
```

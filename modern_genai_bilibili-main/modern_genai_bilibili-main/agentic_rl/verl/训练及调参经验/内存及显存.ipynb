### cpu oom or gpu oom

> 注意区分是 cpu memory (oom_kill, exitcode -9) 还是 gpu memory OOM

- An exit code of "-9" indicates a SIGKILL signal in Linux/UNIX, meaning a process was terminated by the system, often due to insufficient memory and the kernel's "out-of-memory" (OOM) killer
- gpu memory:
    - torch.cuda.OutOfMemoryError

### checkpoint save

- 7B sft 训练一个 epoch 保存 checkpoints 时会有峰值 cpu 内存的占用
    - params: 7B * 2 = 14GB
    - AdamW: 7B * 4 * 2 = 56GB
        - 4: fp32,
        - 2:
            - 一阶矩估计 $m$（这是过去梯度的指数移动平均值。你可以把它理解为梯度的动量（Momentum），它帮助加速参数在正确方向上的更新并抑制振荡。）
            - 二阶矩估计 $v$（这是过去梯度平方的指数移动平均值。你可以把它理解为梯度的方差（Variance），它用于为每个参数自适应地调整学习率。梯度变化大的参数，学习率会相应减小；梯度变化小的参数，学习率会相应增大。）

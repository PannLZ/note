  | 维度 | DeepSeek V3.2 | Kimi K2.5 |
  |---|---|---|
  | 粒度 | sequence-level mask | token-level clipping / masking |
  | 触发条件 | 只 mask negative advantage 且 divergence 大的样本 | ratio / log-ratio 超出区间就处理，不依赖 advantage 符号 |
  | divergence 形式 | sequence mean $\log(\pi_{\text{old}}/\pi_\theta)$ | token-level $\log(\pi_\theta/\pi_{\text{old}})$ |
  | 操作方式 | hard mask：$M_{i,t}=0$ | clip ratio + log-ratio penalty |

  | 目标 | 避免 highly off-policy negative samples 误导训练 | 限制 token-level off-policy drift，尤其训练/推理框架不一致导致的偏移 |

### dpsk v3.2

- Off-Policy Sequence Masking
    - 如果一个 response 是负样本，并且它和当前 policy 的 divergence 太大，就不要用它更新模型。
    - 注意它只 mask negative sequence。原因是：
        - 正 advantage 样本：即使有点 off-policy，也可能仍然是值得模仿的好样本。
        - 负 advantage 样本：如果它本来已经离当前 policy 很远，再强行让模型“远离它”，可能会产生误导性梯度，导致不稳定。

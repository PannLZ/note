- PPO 比着 grpo 更精细，优势定义在single step上（基于每一步的critic）。
    - 但为什么 rl4llm 可行，llm 经过大规模 pretraining 之后其实已足够稳定，初始状态下已经是一个很好的 init policy model 了
    - 反而是传统的 rl tasks，ppo > grpo；

### deepseek v3.2

- without $\sigma$ of adv: $\hat A_{i,t}=R_i-\text{mean}(\mathbf R)$
- unbiased K3 estimator
- off-policy sequence masking
    - off-policy behavior：
        - a **large batch** of rollout data, subsequently split into **multiple mini-batches** for several gradient update steps.
    - training-inference inconsistency
    - To stabilize training and improve tolerance for off-policy updates, we mask negative sequences that introduce significant policy divergence, as measured by the KL divergence between the data-sampling policy $\pi_{old}$ and the current policy $\pi_\theta$.
    - highly off-policy negative samples can be detrimental

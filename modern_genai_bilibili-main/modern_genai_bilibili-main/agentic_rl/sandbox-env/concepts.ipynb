- 并发
    - 并发从个位数到单机两百以上、集群到上千/上万“沙箱单位”都存在，差异主要来自 环境有多“重”（浏览器/OS 比纯代码执行重很多）、启动/重置成本、以及 GPU 推理吞吐 是否被充分喂满。
    - 环境
        - CI：code interpreter
        - 浏览器，OS
- docker 容器
    - retool: `docker run -it -p 8080:8080 volcengine/sandbox-fusion:server-20250609`
        - 它更像是“模型 rollout（生成轨迹） + 共享的代码沙箱服务”的组合。
        - 为了加速 RL 训练，设计了 异步代码沙箱环境，沙箱以 pods（可理解为 K8s Pod/容器实例） 组成 worker pool，按容量拉取任务，异步并行避免慢线程拖累吞吐。
        - 并发更多由“工具调用限流”决定，而不是“Docker 数量=并发数”
            - 即便只起 1 个 SandboxFusion server 容器，也可以在容器内部用多 worker 处理并发；或者横向起多个 server 容器做负载均衡。

### 权重量化 vs. 浮点计算格式

- INT4 是整数权重量化，FP8 是浮点计算格式，用途与数值表示都不同。
    - INT4：4‑bit 有符号两补码的量化表示，需要按块 scale；
    - FP8：8‑bit 浮点格式（例如 E4M3 / E5M2），用于计算加速与动态范围/精度折中。
    - RTX 4090：Ada Lovelace 的 4 代 Tensor Cores 支持 FP8；官方也明确 RTX 4090 有 FP8 Tensor Engine。

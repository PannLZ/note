### AsyncLLMEngine

- This class is used to wrap the LLMEngine class to make it **asynchronous**. It uses **asyncio** to create a background loop that keeps processing incoming requests.
- Continuous Batching vs. offline batching
    - offline batching: 入口是封闭的。一旦 generate 函数开始跑，新的请求进不来，必须等这一批跑完。即通常在 API 设计上，你需要等整个 list 处理完（或者分块处理完），程序才会返回结果。
        - vLLM 的离线推理本质上就是“输入数据一次性全部就绪”的 Continuous Batching。
    - Continuous Batching：没有固定的“批次”概念。
        - 解决 Straggler（落后者/长尾）的问题
- when
    - Offline 模式：注重的是 Throughput (吞吐量)。比如处理数据集，我不关心第一条数据什么时候出结果，我只关心 1000 条数据总共跑多久。
    - Async 模式：注重的是 Latency (延迟)。由于新请求可以随时插入（只要有显存），用户发出请求后，引擎可以立刻安排它进入下一个 Iteration，用户能以最快速度看到第一个字。
        - Async 的优势在于服务化（Serving）。如果你是在做一个 Server（服务器），必须用 Async。因为服务器的请求是随机到达的。

- P50: 第 50 个百分位（50th percentile），也就是中位数。
    - 把一批请求的时延从小到大排序，位于中间位置的那个值就是 P50。表示有 50% 的请求时延不超过这个数。
- TTFT, TTE
    - TTFT：time to first token，首字时延
    - TTE：Time To End / End-to-End latency
        - TTE ≈ TTFT + output_tokens / T_decode

- `--no-cache-dir`
    - 禁止 pip 使用或保存缓存数据。
    - 默认情况下，pip 会将下载的 .whl 文件（Wheel包）保存在本地（通常在 ~/.cache/pip），以便下次安装同一版本时直接使用，无需重新下载。
    - 使用了 --no-cache-dir 后：
        - 强制重新下载： pip 会忽略本地可能已经存在的包，直接从 PyPI 或指定源重新下载文件。
        - 不留垃圾： 下载并安装完成后，pip 不会保留下载的 .whl 源文件，从而节省磁盘空间。

### cuda

- nvidia-smi 12.1 的驱动，可以跑 torch 2.8.0+cu128 的 gpu 程序
- `export CUDA_MODULE_LOADING=LAZY`
    - 它会延迟 CUDA 模块的加载，直到真正需要时才初始化 CUDA（`torch.cuda.is_initialized()`）。
    - `import torch`/`torch.cuda.is_available()`均不会触发 cuda 初始化
    - `from torch.distributed.device_mesh import init_device_mesh`: 会触发初始化

```sh
python -c "import torch; print(torch.__version__)"
python -c "import torch; print(torch.version.cuda)"
python -c "import torch; print(torch.cuda.get_arch_list())"
```

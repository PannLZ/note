### dqn

- 离散有限状态空间，Q table
- 连续海量状态空间，Deep Q Network，DQN 算法（参数化的函数 $Q_\theta$）便可以用来解决连续状态下离散动作问题；


$$Q(s, a) \leftarrow Q(s, a) + \alpha \left[ r + \gamma \max_{a' \in A} Q(s', a') - Q(s, a) \right]$$


> dqn 中的 policy

- DQN的最终目标是导出一个隐式的、确定性的贪心策略 ($\pi(a|s)=\arg\max_a Q(s,a)$)。
- 在训练过程中，DQN会使用一个显式的、随机性的行为策略（如ε-greedy）来保证充分的探索。

### dqn vs. ppo

- DQN (基于价值的算法 - Value-Based):
    - DQN的目标是学习一个“价值评估函数”（Q函数），这个函数能告诉机器人在任何一个位置（状态 s），采取每个可能的动作（动作 a）能够带来的未来总回报的期望值有多高。
    - 机器人站在迷宫的某个路口，它会利用这个Q函数评估向左、向右、向前走分别有多“值钱”，然后选择那个价值最高的动作。它学习的不是“策略”本身，而是评估每个动作的“价值”。
    - 离策略 (Off-Policy)：可以使用“旧”的经验数据进行学习。
        - DQN学习的是Q值——一个动作的客观价值。无论智能体当时是出于“新手的好奇心”还是“专家的判断”做出这个动作，这个动作最终带来的结果（奖励）是客观存在的。因此，用过去的经验来更新对动作价值的判断是完全合理的。这就好比，你可以通过阅读一本很老的食谱（旧的经验）来学习一道菜的好坏（价值），这并不会有什么问题。
- PPO (基于策略的算法 - Policy-Based):
    - PPO的目标是直接学习一个“策略函数”（Policy），这个函数会直接告诉机器人在某个位置（状态 s）应该以多大的概率选择哪个动作（动作 a）。
        - PPO作为一种策略梯度算法，它更新策略的依据是：“如果我这次采取的某个动作得到了比预期更好的奖励，那么我就应该增加以后在类似情况下采取这个动作的概率”。
        - 这里的关键在于“比预期更好”，这个“预期”是基于当前策略的。
    - 机器人站在迷宫的路口，策略函数会直接给出一个行动指南，比如“70%的概率向前走，20%的概率向左，10%的概率向右”。它直接学习在特定情境下该怎么行动。
    - 在策略 (On-Policy)：只能使用“当前”策略产生的数据进行学习。

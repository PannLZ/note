### MDP

- 如何判断一个环境是否是 mdp 的
    - 在标准的 MDP（马尔可夫决策过程）定义中，有一个隐含的**“同步”和“原子性”**假设：世界是回合制（Turn-based）的
    - 状态转移的无记忆性
        - 是 MDP： 一个抛物线运动的小球，如果“状态”定义为 (位置, 速度)，那么下一秒它在哪只取决于当前的 (位置, 速度)，不需要知道它之前是从哪飞过来的。
        - 非 MDP： 一个小球，如果“状态”只定义为 (位置)。为了预测下一秒，你必须知道它上一秒的位置来推算速度。因为现在的状态缺少了“速度”信息，必须依赖历史。    

### model-based vs. model free

- model-based vs. model free
    - model-based：学习/使用环境的 dynamics $p(s'|s,a)$
    - model free: 直接从经验中学习 $\pi(a|s)$ 或价值函数 $V,Q$
        - 大致可以理解为 policy-based

- VIP（Value-Implicit Pre-training） 范式提出了一种革命性的观点：**人类完成任务的视频本身就是一条价值递增的轨迹**。通过学习这种时间上的平滑性，VFM 可以在没有任何显式奖励标签的情况下，构建出指导机器人动作的势能场。
    - 人类在视频中执行任务的过程，本质上是一个从低价值状态向高价值状态优化的过程。
    - 对于视频序列中的帧 $o_t$ 和 $o_{t+k}$（其中 $k>0$），训练目标使得：
        - $\| \phi(o_t) - \phi(o_g) \|_2 > \| \phi(o_{t+k}) - \phi(o_g) \|_2$
- Value head
    - 基于一个预训练的 LLM（如 Llama-3-8B 或 Nemotron-340B-Base），移除其原本的语言建模头（Unembedding Layer），取而代之的是一个随机初始化的线性层（Linear Head）。该层将 Transformer 输出的最后一个 Token 的隐状态（Hidden State）映射为一个标量值。
    - 这种架构能够最大程度地**保留基座模型的语义理解能力**。

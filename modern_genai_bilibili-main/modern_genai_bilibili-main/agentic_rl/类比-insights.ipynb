### 数学推理能力

- “数学推理能力”不是一条线，而是一个 d 维（比如 d=20）的空间，每一维代表一种相对独立的能力方向（竞赛解题、统计、优化、概率、泛函、ML 理论、数值分析……）。
    - 每次你用某类数据（比如 Codeforces/ICPC/IMO）去做 finetune / RL，本质上只是在这个 20 维球中，某个局部区域“打补丁”（能力大幅提升），但只在附近的任务上泛化。
    - 想依靠这种“一个任务 → 一个小补丁”的方式把整个“数学能力空间”都铺满，会遇到维度灾难：需要的补丁数量随维度指数级爆炸 → 基本不可能。
- 用半径为 r 的小球（或小块）去覆盖一个半径为 1 的 d 维球，需要大约多少个小球？
    - 如果 $r=0.5$, $N(r)=2^{20}=10^6$

$$
N(r) \approx \left(\frac{R}{r}\right)^d.
$$

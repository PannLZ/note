### DeepResearch Bench

- DeepResearch Bench（100个博士级任务）以及两套全新的评估框架：RACE（评估报告质量）和 FACT（评估引用真实性）
- FACT (Factual Abundance and Citation Trustworthiness) 关注 Agent 检索信息的真实性。它不依赖于文本生成质量，而是检查“每一个引用是否支持其对应的陈述”。

- 动态权重生成 (Dynamic Weighting): Judge LLM (Gemini 2.5 Pro) runs $T$ trials
    1. Comprehensiveness (Information coverage)
    2. Insight (Depth, originality)
    3. Instruction Following
    4. Readability
- 为了避免 LLM 打分时的“分数膨胀”（倾向于给高分），RACE 采用相对评分法。它将待测报告与一个“标杆”进行对比。

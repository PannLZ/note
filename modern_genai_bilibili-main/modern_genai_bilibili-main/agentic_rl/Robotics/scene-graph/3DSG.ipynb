$$
\mathcal G=(\mathcal V, \mathcal E)
$$

- $\mathcal V=\left\{v_1, v_2, \cdots, v_n\right\}$  代表 3D 物体节点，每个节点 $v_i=\left\{p_i, s_i, \mathcal A_i, \mathcal C_i\right\}$
    - $p_i$：位置，$s_i$：尺寸，$\mathcal A_i$ 属性集合 (颜色、材质、形状), $\mathcal C_i$：类别标签
 
| Category | Relation |
| :--- | :--- |
| **Vertical Proximity** | support, hanging on, mounted on, below, higher than, embed, inside, affixed on, above, lower than |
| **Horizontal Proximity** | near, besides, far, next to |
| **Allocentric** | left, behind, right, is in front of |


- Vertical Proximity：包含了垂直空间及接触关系的词汇（如 support, above 等）。
- Horizontal Proximity：包含了水平距离相关的词汇（如 near, next to 等）。
- Allocentric：包含了以物体为中心的方向词汇（如 left, behind 等）。

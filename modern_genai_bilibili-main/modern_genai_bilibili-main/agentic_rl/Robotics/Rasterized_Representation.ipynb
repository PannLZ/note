### 2d Rasterized Image

- 空间结构：哪里是墙（不可通行），哪里是空地（可通行），哪里是未知区域。
- 轨迹 (Trajectory)：机器人从开始到现在的移动路径。
- 地标 (Landmarks)：这是最关键的部分。地图上会标注具体的 数字或符号（Points of Interest），对应于：
    - 探测到的关键物体（如 "Sofa", "Door"）。
    - 房间的采样中心点。
    - 关键的观测位置。

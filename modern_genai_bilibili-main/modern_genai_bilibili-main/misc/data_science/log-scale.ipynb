$$P_{log}(x) = \log_{10}(x)$$
- 首先大的间距
    - $\log(10)-\log(1)=\log\frac{10}{1}=1$
    - $\log(1)-\log(0.1)=\log\frac{1}{0.1}=1$
- $\Delta d_{1\to2} = \log_{10}\left(\frac{2}{1}\right) = \log_{10}(2) \approx \mathbf{0.301}$
- $\Delta d_{8\to9} = \log_{10}\left(\frac{9}{8}\right) = \log_{10}(1.125) \approx \mathbf{0.051}$
- 刻度数量
    - 1-10：2, 3, 4, 5, 6, 7, 8, 9。9段；

```python
import matplotlib.pyplot as plt
import numpy as np

# 设置画图风格
plt.style.use('dark_background')
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 6), constrained_layout=True)

# 数据点：从 1 到 10
x_ticks = np.arange(1, 11)

# === 第一张图：Linear Scale (English) ===
ax1.set_xlim(0.5, 10.5)
ax1.set_ylim(0, 1)
ax1.set_yticks([])
ax1.set_xticks(x_ticks)
ax1.spines['top'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.spines['right'].set_visible(False)
# 英文标题：Linear Scale - Focus on Absolute Difference
ax1.set_title("Linear Scale: Focus on 'Addition' (Absolute Diff)", fontsize=14, color='white', pad=20)

# 画刻度线
for x in x_ticks:
    ax1.vlines(x, 0, 0.2, color='white')

# 高亮 1-2 区间
ax1.axvspan(1, 2, color='cyan', alpha=0.3)
ax1.text(1.5, 0.5, "1 -> 2\n(+1 Unit)", color='cyan', ha='center', fontweight='bold')

# 高亮 8-9 区间
ax1.axvspan(8, 9, color='yellow', alpha=0.3)
ax1.text(8.5, 0.5, "8 -> 9\n(+1 Unit)", color='yellow', ha='center', fontweight='bold')

# === 第二张图：Log Scale (English) ===
ax2.set_xscale('log')
ax2.set_xlim(0.9, 10.5)
ax2.set_ylim(0, 1)
ax2.set_yticks([])
ax2.set_xticks(x_ticks)
ax2.set_xticklabels([str(x) for x in x_ticks])
ax2.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, _: str(int(x))))
ax2.spines['top'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines['right'].set_visible(False)
# 英文标题：Log Scale - Focus on Multiplication/Ratio
ax2.set_title("Log Scale: Focus on 'Multiplication' (Ratio/Growth)", fontsize=14, color='white', pad=20)

# 画刻度线
for x in x_ticks:
    ax2.vlines(x, 0, 0.2, color='white')

# 高亮 1-2 区间
ax2.axvspan(1, 2, color='cyan', alpha=0.3)
ax2.text(1.414, 0.5, "1 -> 2\n(x2 Double)", color='cyan', ha='center', fontweight='bold')

# 高亮 8-9 区间
ax2.axvspan(8, 9, color='yellow', alpha=0.3)
ax2.text(8.48, 0.5, "8 -> 9\n(x1.125 Growth)", color='yellow', ha='center', fontweight='bold')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Prepare Data
# Generate points from 0.1 to 100
x = np.linspace(0.1, 100, 500)
# The Function: y = 50 * log10(x) + 100
# This simulates "Diminishing Returns": Input x10 -> Score +50
y = 50 * np.log10(x) + 100

# 2. Setup Plot Style
plt.style.use('dark_background')
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), constrained_layout=True)

# === Figure 1: Linear Scale ===
ax1.plot(x, y, color='#00ff00', linewidth=2)
ax1.set_title("Figure 1: Linear Scale (Standard Ruler)\nThe curve flattens out: 'Diminishing Returns'", fontsize=14, color='white')
ax1.set_ylabel("Performance Score")
ax1.grid(True, alpha=0.3)

# Mark key points
ax1.scatter([0.1, 1, 10, 100], [50, 100, 150, 200], color='red', zorder=5)
ax1.text(10, 140, '  x=10\n  y=150', color='white', verticalalignment='top')
ax1.text(100, 190, '  x=100\n  y=200', color='white', verticalalignment='top')

# === Figure 2: Log Scale ===
ax2.plot(x, y, color='#00aaff', linewidth=2)
ax2.set_xscale('log')  # <--- The Magic: Logarithmic X-Axis
ax2.set_title("Figure 2: Log Scale (Ratio Ruler)\nThe curve becomes a STRAIGHT LINE", fontsize=14, color='white')
ax2.set_xlabel("Cost / Input ($)")
ax2.set_ylabel("Performance Score")
ax2.grid(True, which="both", alpha=0.3)

# Force showing key ticks as plain numbers (0.1, 1, 10...)
ax2.set_xticks([0.1, 1, 10, 100])
ax2.set_xticklabels(['0.1', '1', '10', '100'])

# Draw 3 equal-width regions to prove the "Ratio Consistency"
# Region A: 0.1 -> 1
ax2.axvspan(0.1, 1, color='yellow', alpha=0.15)
ax2.text(0.3, 120, "Region A\n(10x Growth)\n$0.1 -> $1", color='yellow', ha='center', fontweight='bold')

# Region B: 1 -> 10
ax2.axvspan(1, 10, color='magenta', alpha=0.15)
ax2.text(3, 170, "Region B\n(10x Growth)\n$1 -> $10", color='magenta', ha='center', fontweight='bold')

# Region C: 10 -> 100
ax2.axvspan(10, 100, color='cyan', alpha=0.15)
ax2.text(30, 220, "Region C\n(10x Growth)\n$10 -> $100", color='cyan', ha='center', fontweight='bold')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data representing prices: $0.1, $1, $10
prices = np.array([0.1, 1, 10])

plt.figure(figsize=(10, 6))

# Plot 1: What the chart actually is (Log Scale Axis)
plt.subplot(2, 1, 1)
plt.scatter(prices, [1, 1, 1], c='blue', s=100)
plt.xscale('log')
plt.title("Scenario A (Correct): Log Axis, Raw Labels\nThe axis does the math, but shows you the original money ($)", fontsize=12)
plt.yticks([])
plt.xlabel("Price ($)")
# Force the ticks to show the raw numbers
plt.xticks([0.1, 1, 10], ['0.1', '1', '10'])
plt.grid(True, which="both", alpha=0.3)

# Plot 2: What the user is worried about (Log Transformed Data)
plt.subplot(2, 1, 2)
log_prices = np.log10(prices) # result is -1, 0, 1
plt.scatter(log_prices, [1, 1, 1], c='red', s=100)
plt.title("Scenario B (Your Worry): Log Transformed Data on Linear Axis\nThe labels would be -1, 0, 1 (Not dollars anymore)", fontsize=12)
plt.yticks([])
plt.xlabel("log10(Price)")
plt.xticks([-1, 0, 1], ['-1 (means 0.1)', '0 (means 1)', '1 (means 10)'])
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('log_explanation.png')
```

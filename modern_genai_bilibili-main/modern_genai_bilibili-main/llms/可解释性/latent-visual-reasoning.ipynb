    - 通过对激活值训练简单的线性分类器（Linear Probes），可以近乎完美地预测当前推理正处于哪个阶段。这意味着大模型并非在“盲目吐字”，其内部隐空间存在着明确的进度导航机制。
    - Mid-reasoning Prediction）：正确与错误的推理在早期的几何轨迹高度重合（如同两人同走一条正确的起始道路），但在后期的步间转移（Between-step Transitions）中，其激活值空间距离（计算为 $\Delta(I-C)$，包括欧氏距离与余弦距离）会产生系统性的显著分化。基于这种晚期阶段的激活值几何特征，分类器能够在最终答案输出前，以高达 0.87 的 ROC-AUC 预测该推理路径的最终正确性。
    - 从观察走向因果控制（Trajectory-based Steering）：激活值不仅能“看”，还能“拨”。通过向特定隐层 $h_t^{(l)}$ 注入微小的方向扰动向量 $\alpha \cdot s^{(l)}$，可以在推理偏离时将其“拽回”理想轨迹，甚至可以通过朝向或背离“终止子空间”（Termination Subspace）施加干预，直接控制模型思考的长短（加速收敛或延长反思）。

    - Progressive Disentanglement:在非常浅的神经网络层（如 Layer 0），所有步骤的激活值混沌地交织在一起（Entangled）；但随着网络深度增加（推进至 Layer 31），原本混合的表征逐渐解开，各自坍缩成界线分明、高度簇集的步骤专属区域。
    - Layer-wise Separability：图 1(b) 的定量分析揭示，早期的简单步骤（Step 1, Step 2）在极浅层就能被线性剥离分离（准确率瞬间逼近 1.0）；而中晚期的深层推理（Step 3, 4, 5）则需要经过更深的 Transformer 模块进行高度非线性变换，才能在更深层实现线性可分。
    - Base vs. Instruct vs. R1-Distill：这是极其反直觉的洞见——这种分步的几何结构在完全未经过指令微调的Base（基座）模型中就已经天然存在。后续的推理强化训练（如 R1-Distill 蒸馏）并没有从零“无中生有”地创造新的思维表示结构，而是加速了这种收敛过程。例如，R1-Distill 模型在极浅层（Layer 0）就已经能够让代表“终止输出”的表征高度独立分离，极大地改变了表征空间组织的时间截面。

    - 前向锚定初期（Early Phase）：统一抽取所有问题从 Step 1 到 Step 2 的转移轨迹（Step 1 $\rightarrow$ 2），以此证明在任何问题的起步阶段，正确与错误的轨迹是无法区分的。
    - 后向锚定末期（Late Phase）：统一抽取所有问题的倒数第二步到倒数第一步（2nd-last $\rightarrow$ Last），以及**倒数第一步到答案标记（Last $\rightarrow$ Ans. Marker）**的转移轨迹 。

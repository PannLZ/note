### lookup table vs. qkv

$$
\begin{aligned}
Q &= (x_{embed} + P) \cdot W_Q \\
K &= (x_{embed} + P) \cdot W_K \\
V &= (x_{embed} + P) \cdot W_V
\end{aligned}
$$

$$
\begin{aligned}
q_m &= \text{RoPE}(x_m \cdot W_Q, m) \\
k_m &= \text{RoPE}(x_m \cdot W_K, m) \\
v_m &= x_m \cdot W_V
\end{aligned}
$$

- Vocabulary Embedding（词嵌入）是 Q、K、V 的“原材料”，而 Q、K、V 是通过“线性投影（Linear Projection）”对这个原材料加工后的产物。
    - RoPE 版本，$m$ 刻画的是 token 所谓的 position
- 多头注意力 (Multi-Head Attention)
    - 在实际的 LLM 中，这个过程会并行发生多次。比如有 8 个头，就会有 8 组不同的 $(W_Q^i, W_K^i, W_V^i)$。这意味着同一个 Embedding 会被投影成 8 组不同的 Q、K、V，分别去捕捉不同的特征（有的头看语法，有的头看指代，有的头看语义）。
- 深层网络的变化上述关系严格适用于第一层（Embedding Layer 之后的那一层）。到了 Transformer 的第 2 层及以后，Q、K、V 的输入来源（那个 $x$）就不再是原始的 Vocabulary Embedding 了，而是上一层的输出（经过了混合和处理后的向量）。但“乘以 $W$ 矩阵进行投影”这个操作机制是一模一样的。

> engram: enhenced gram??

- 现有的 LLM (如 MoE) 虽然通过条件计算（Conditional Computation）扩展了容量，但它们本质上仍然是在用“计算”来模拟“检索”。
    - 例如，要预测 "Alexander the Great"，模型需要消耗多层 Attention 和 FFN 来逐步构建这个实体的表示。这是一种对计算资源的浪费——因为这些固定的知识（Knowledge）本可以通过查表（Lookup）瞬间获得。
- Engram 将底层网络从繁琐的“静态模式重建”中解放出来，使 Attention 能够专注于全局上下文和复杂逻辑。
- Infrastructure-Aware Efficiency
    - Engram 的检索是**确定性**的（Deterministic）。这意味着我们可以在 GPU 计算当前层之前，利用 CPU 异步 **Prefetch**（预取）下一层的 Memory Embedding。

### merge rank

$$
[\text{马},\text{嘉},\text{祺}]\rightarrow[\text{马},\text{嘉祺}]
$$

- merge rank 可以理解为 BPE tokenizer 里的“合并优先级”。
- 每一次合并都会被记录下来，形成一个有顺序的 merges 表。这个顺序就是 merge rank。通常：
    - rank 越小⇒合并越早学到⇒encode 时优先级越高
- `马嘉祺 -> [马, 嘉祺]` 的 encode 过程
    - 第一轮看所有相邻 pair：$(\text{马}, \text{嘉}),\quad(\text{嘉}, \text{祺})$
        - $(\text{马}, \text{嘉}) \notin \text{merges}$
        - $(\text{嘉}, \text{祺}) \in \text{merges}$
        - 所以执行唯一可用合并：$[\text{马}, \text{嘉}, \text{祺}]\rightarrow[\text{马}, \text{嘉祺}]$
    - 第二轮再看当前序列里的相邻 pair：
        - $(\text{马}, \text{嘉祺})$，
        - $(\text{马}, \text{嘉祺}) \notin \text{merges}$那算法停止。最终结果就是：$\text{马嘉祺}\rightarrow[\text{马}, \text{嘉祺}]$
- 如果没有 嘉祺，但有 马嘉祺
    - 如果词表里只有 马嘉祺，同时没有 嘉祺、没有 马嘉，那在字符级 BPE 抽象下，通常无法从 `[马, 嘉, 祺]` 合并到 `[马嘉祺]`。因为 BPE 每一步只能合并相邻的两个片段，不能一步把三个片段直接合成一个。

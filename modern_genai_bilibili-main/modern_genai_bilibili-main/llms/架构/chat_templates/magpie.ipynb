- Pseudo Experience Replay / 伪经验回放
- BOS + user prefix + assistant prefix
    - Magpie-style synthetic sample：也就是模型根据模板前缀“想象”出一个用户问题，再给出答案。

$$
q_{\text{Magpie}}(x,y)
=p_\theta(x\mid \tau_U)\;
p_\theta(y\mid \tau_U \oplus x \oplus \tau_A)
\approx
p_{\text{SFT}}(x,y\mid T_{\text{pre\_user}})
$$

- $x$ 是用户问题，$y$ 是助手回答，$\tau_U$ 是“用户消息开始前”的模板前缀，$\tau_A$ 是“助手消息开始前”的模板前缀，$\oplus$ 表示字符串或 token 序列拼接。
    - $\tau_0 =\texttt{<bos><system>...<user>}, \tau_1 =\texttt{<assistant>}$
    - 后训练数据可以看成一批被模板包起来的序列：$z=\tau_0\oplus x\oplus \tau_1\oplus y\oplus \tau_2$
        - $z
=\underbrace{\tau_0}_{\text{BOS + system + user prefix}}
\oplus\underbrace{x}_{\text{user content}}
\oplus\underbrace{\tau_1}_{\text{assistant prefix}}
\oplus\underbrace{y}_{\text{assistant content}}
\oplus\underbrace{\tau_2}_{\text{EOS / turn end}}$
    - $z=\texttt{<bos><system>}\oplus s\oplus\texttt{<user>}\oplus x\oplus\texttt{<assistant>}\oplus y\oplus\texttt{<eos>}$
    - $\text{chat template}=(\tau_0,\tau_1,\tau_2)$

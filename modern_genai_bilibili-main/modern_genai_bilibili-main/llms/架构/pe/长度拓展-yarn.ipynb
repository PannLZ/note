在基于 RoPE 进行 Context Window（上下文窗口）拓展时，主要依赖于**“位置插值（Position Interpolation）”与“基频调整（Base Frequency Scaling）”相结合的机制，并配合长文本持续预训练**。不是让模型去“预测”没见过的更远的位置（外推），而是通过数学变换，把更长的位置“压缩”进模型原本熟悉的范围内（内插），同时调整旋转频率以保持分辨率。

- 直接外推（Extrapolation），即直接给模型输入超过训练长度的位置索引（如训练时最大是 4096，现在输入 5000）。但这在 RoPE 上效果极差，因为 RoPE 的旋转角度在超出训练范围后会变得极其混乱（Out-of-Distribution, OOD），导致 Attention 机制崩溃。

**空域（Spatial Domain）关注的是像素在二维空间中的物理排布与颜色深浅；而频域（Frequency Domain）**则是将整幅图像拆解为无数个不同方向、不同频率的二维正弦/余弦波的叠加。


$$
\mathcal F(I)\rightarrow \text{Mag}, \text{Phase}
$$

- 将空域图像通过傅里叶变换进入频域：
- 对数幅度谱 $\log(f)=\log(\text{Mag})$
    - 中间极亮（低频能量高），边缘暗（高频能量低），这是自然图像的绝对规律。
        - 像素的亮度（Brightness）是其对应频率分量幅值（Magnitude）的视觉量化，而幅值的平方即代表该频率的能量（Energy）。
    - 信号处理界一个著名的自然图像统计学定律——$1/f$ 幂律分布（$1/f$ Power Law）。自然图像的能量随着频率 $f$ 的升高，大致呈现 $1/f^\alpha$ （$\alpha \approx 1$）的指数级衰减。

$$
\text{Energy}(u,v) \propto |F(u,v)|^2
$$

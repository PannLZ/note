- sam2: PVS (promptable visual segmentation)
    - points：
        - label 1 indicates a positive click (to add a region) while label 0 indicates a negative click (to remove a region).
            - https://colab.research.google.com/github/facebookresearch/sam2/blob/main/notebooks/video_predictor_example.ipynb
    - bbox:
        - grounding sam2
    - 根据用户的视觉提示（点、框、掩膜）来分割**单个目标对象**。
        - 对象级 (Instance-level): 每次交互只对应一个特定的物体。即使画面中有多个相同的物体，PVS 也只分割你指定的那个。
        - 语义无关 (Agnostic): 模型不需要“懂”这个物体是什么（比如它不需要知道这是“猫”），只需要根据几何提示把这个视觉上的连通区域抠出来。
- sam3: PCS (Promptable Concept Segmentation)
    - 根据“概念提示”（文本名词短语、图像范例或二者结合），在图像或视频中检测、分割并追踪所有符合该概念的实例。
    - 类别级/概念级 (Concept-level): 关注的是“一类事物”。一次输入可以分割出多个实例（Everything like this）。
    - 语义感知 (Semantic-aware): 模型必须理解提示中的语义（例如理解什么是 "red apple"），或者理解图像范例的视觉特征，从而在全图中寻找匹配项。
    - 多模态输入: 支持 文本（Text）、图像范例（Image Exemplar）以及 文本+范例 的混合输入。

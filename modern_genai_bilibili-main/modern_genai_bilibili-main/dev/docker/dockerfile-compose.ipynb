一个复杂的软件系统通常不由一个程序组成，而是由一堆程序配合工作（前端、后端、数据库、缓存...）。

- `docker compose up` 时，它就会乖乖地把这些全部拉起来。
    - `compose.yml`
        - `services`
            - `frontend`
            - `backend`
        - `volumes`

### compse.yml

```yml
services:
  frontend:
    network_mode: host
  backend:
    network_mode: host
  redis-for-backend:
    network_mode: host
```

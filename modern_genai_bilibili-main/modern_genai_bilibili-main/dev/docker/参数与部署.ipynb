- `--net=host`
    - docker run --net=host（或 --network host）的含义是：让容器直接使用宿主机（Host）的网络堆栈，取消容器与宿主机之间的网络隔离。
    - 这意味着容器不会拥有自己独立的 IP 地址，而是直接共用宿主机的 IP 和端口。
- `docker run --gpus all ubuntu nvidia-smi`
    - 检测 docker/nvidia 环境是否正常

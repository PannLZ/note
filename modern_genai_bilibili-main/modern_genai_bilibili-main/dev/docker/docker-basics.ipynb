### misc 

- `xhost +` 既不是 Docker 的命令，也不是环境变量。
    - 它是 Linux 系统自带的一个原生系统命令，属于 X Window System（X11，Linux 下的图形界面底层系统）的一部分。
    - 允许 Docker 容器访问宿主机的 X server，这通常是为了支持仿真环境的 3D 渲染或可视化界面的显示。

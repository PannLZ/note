### heapq

- `heapq.heapify`: in-place 堆化（默认最小堆）
    - x[0]（根节点）永远是列表中的最小值。
    - 对于堆中任意索引 $k$ 的节点，其值都小于或等于其子节点的值。
        - 索引 $k$ 的父节点是 $(k-1)//2$
        - 索引 $k$ 的左子节点是 $2k + 1$
        - 索引 $k$ 的右子节点是 $2k + 2$
- `heapq.heapreplace`

```python
import heapq
data = [3, 1, 4, 1, 5, 9, 2, 6]
heapq.heapify(data)
data
```

```python
k = 2
assert data[(k-1) // 2] <= data[k]
assert data[k] <= data[2*k+1]
assert data[k] <= data[2*k+2]
```

```python
# 1. 初始化 Worker 池
# 格式: (worker_available_time, worker_name)
# 初始时，所有 worker 都在时间 0 空闲
workers = [
    (0, 'Worker A'),
    (0, 'Worker B'),
    (0, 'Worker C')
]

# 将列表转换为最小堆
heapq.heapify(workers)
```

```python
print(f"初始状态 (堆): {workers}\n")
```

```python
# 2. 模拟任务流 (任务耗时)
# 假设来了 6 个任务，每个任务需要花费的时间如下
tasks_duration = [10, 30, 5, 20, 10, 50]
print("--- 开始任务分配 ---")
```

```python
for i, duration in enumerate(tasks_duration):
    task_id = f"Task-{i+1}"
    
    # 核心操作：heapreplace
    # 我们不做 pop 再 push，而是直接查看堆顶（最早空闲的人），
    # 算出他做完这个新任务后的新时间，然后“替换”他在堆里的状态。
    
    # 1. 获取当前堆顶信息（最小的可用时间）
    current_available_time, worker_name = workers[0]
    
    # 2. 计算该 worker 做完新任务后的新空闲时间
    new_available_time = current_available_time + duration
    
    print(f"分配 {task_id} (耗时 {duration}s) -> 给 {worker_name} "
          f"(原空闲: {current_available_time}, 新空闲: {new_available_time})")
    
    # 3. 使用 heapreplace 更新堆
    # 语义：这个 worker 原来的空闲状态结束了（被 Pop），
    # 变成了新的空闲状态（被 Push），堆维持固定大小。
    heapq.heapreplace(workers, (new_available_time, worker_name))
    
    # 打印当前堆的状态（注意：堆的物理存储顺序不一定完全有序，但 workers[0] 永远是最小）
    print(f"    当前堆状态: {workers}")

print("\n--- 最终结果 ---")
# 把堆里的东西都弹出来看看最终谁最晚下班
while workers:
    print(heapq.heappop(workers))
```

- detached head
    - 任何将 HEAD 指针直接指向具体的 Commit ID（提交哈希值） 而不是指向分支名（Branch Name） 的操作，都会导致 "detached HEAD"（HEAD 游离）状态。

### git pull 

- git pull: 默认会有一个 merge 的 commit（如果当前分支远端已有提交的话）
    - 会导致大量的没有信息量的 commit history
    - `git config pull.rebase true`：保持 linear 的提交历史

### 回退

- git reset --hard
    - 永久删除提交，改变历史
- git revert
    - 创建新提交，保留历史
- git checkout
    - 临时切换，不改变历史，查看专用
    - 会导致 detached head

- 静态语法、代码风格检测（pre-commit）和统一
    - `.pre-commit-config.yaml`
    - `pre-commit`
        - `pre-commit --version`
    - 当运行 git commit 时：
        - Git 触发 .git/hooks/pre-commit 脚本。
        - Pre-commit 读取 .pre-commit-config.yaml。
        - 会去下载 repo 中指定的仓库
        - 在隔离环境中执行 hooks 下定义的命令（如 black, ruff）。

- Black/Flake8/Ruff/Pylint
    - black：排版 (空格, 换行)，自动修复
        - `uv run black xx.py`
    - flake8：语法 + 基础风格，否 (只报错)
    - Ruff：语法 + 风格 (全能)，是 (可自动修复)
        - `ruff check --fix xx.py` 手动命令修复
    - Pylint：逻辑 + 质量 + 复杂度，否（只报错）
        - too-many-branches
        - `(uv run) pylint `

- streamlit：脚本即应用 (Script-based)
    - 像写 Python 脚本一样写网页。
    - 更适合数据仪表盘（Dashboard）
- gradio
    - 更适合 ai 的前端展示
    - 定义接口：输入 -> 函数 -> 输出
        - `gr.Interface(fn=to_upper, inputs="text", outputs="text")`

### demos

- `streamlit run streamlit_demo.py`

```python
import streamlit as st

st.title("大写转换器")
# 脚本从这里开始执行，每次交互都从这里重跑
user_input = st.text_input("请输入文字")

if st.button("转换"):
    st.write(user_input.upper())
```

- `python gradio_demo.py`

```python
import gradio as gr

def to_upper(text):
    return text.upper()

# 定义接口：输入 -> 函数 -> 输出
demo = gr.Interface(fn=to_upper, inputs="text", outputs="text")
demo.launch()
```

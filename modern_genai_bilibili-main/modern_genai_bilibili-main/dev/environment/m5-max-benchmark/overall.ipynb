### ane (apple NPU) vs. GPU

- Apple M 芯片里的 GPU，本质上是“面向图形与通用并行计算的大规模并行处理器”；ANE，本质上是“面向神经网络推理的专用加速器”。
- ANE 的硬件底层是一种类似于脉动阵列（Systolic Array）的极度定制化结构。如果它像 GPU 那样高频、宽泛地直接去外部的 LPDDR 主存中读写每一层网络的数据，极高的物理延迟与数据搬运的功耗墙将瞬间摧毁其引以为傲的“毫瓦级能效比”。
- 因此，在 ANE 的硅片内部，紧紧贴着其乘加计算单元（MACs），工程团队极其奢侈地安置了一块容量极小（通常在数 MB 到十几 MB 量级）但数据吞吐率极高的专属片上静态随机存取存储器（SRAM）。

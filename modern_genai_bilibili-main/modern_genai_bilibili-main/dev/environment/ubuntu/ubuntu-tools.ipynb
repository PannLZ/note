- iterm shell 可以通过 安装 `imgcat` 的方式，在命令行显示 img

```
# 1. 下载脚本
curl -L https://iterm2.com/utilities/imgcat > imgcat

# 2. 赋予执行权限
chmod +x imgcat

# 3. 移动到系统路径（这样你在任何目录都能直接输命令）
sudo mv imgcat /usr/local/bin/
```

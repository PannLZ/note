### best practices

- 用 rsync 而非 scp
    - `rsync -av test/src test/dst/`
        - -a：archive，尽量保留源文件的各种属性，并递归同步目录。它大致等价于一组常用选项的合集：
        - -v（verbose，详细输出）
        - --delete：让目标端删除源端已不存在的文件（做“镜像同步”时用，慎用）
        - rsync 的好处主要在“同步”而不是“复制”：
            - 增量传输（省时间/省流量）：只传有变化的部分（默认按大小/时间判断，必要时可用校验），大文件改一点点也不用重传整文件。
        - 注意如果保留目录，原路径不能加 `/`，即需要是 `test/src` 这种形式

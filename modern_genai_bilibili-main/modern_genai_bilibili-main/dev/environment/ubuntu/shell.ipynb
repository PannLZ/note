- `"$@"`：外部位置参数的完美透传
    - 概念本质：`$@` 是一个特殊的 Bash 变量，代表着调用当前这个外层脚本时，外部传入的所有参数列表。

```shell
BASE_ARGS="--model gpt-4 --temperature 0.2 --max-tokens 1024"
python run.py $BASE_ARGS --input data.json

# shell 数组
BASE_ARGS=(--model gpt-4 --temperature 0.2 --name "John Smith")
python run.py "${BASE_ARGS[@]}"

python run.py --model gpt-4 --temperature 0.2 --name "John Smith"
```
